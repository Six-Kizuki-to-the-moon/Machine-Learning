```python
import googlemaps
import pandas as pd
import time
```

```python
cities_jakarta = ['Jakarta Selatan', 'Jakarta Utara', 
                  'Jakarta Timur', 'Jakarta Barat', 'Jakarta Pusat']
```

```python
# Initialize the Google Maps client with your API key
api_key = 'AIzaSyBr7aAxeDamZx5L3ENlk9AW1dQHrSNS1ZM'
gmaps = googlemaps.Client(api_key)
df = pd.DataFrame(columns=['Name', 'Rating', 'Address', 'Location'])
def scrape_hotels_in_city(df, gmaps, city):

    # Define the search query
    query = 'hotel'

    # Perform a text search for hotels in the specified city
    places_result = gmaps.places(query, location=city)

    # Extract relevant data from the initial API response
    places = places_result['results']
    data = []
    for place in places:
        name = place['name']
        rating = place.get('rating', 'N/A')
        address = place['formatted_address']
        location = place['geometry']['location']
        data.append({'Name': name, 'Rating': rating, 'Address': address, 'Location': location})
        time.sleep(2)

    # Check if there are more results available
    for i in range(2):
        if 'next_page_token' not in places_result:
            break

        places_result = gmaps.places(query, location=city, page_token=places_result['next_page_token'])
        places = places_result['results']
        for place in places:
            name = place['name']
            rating = place.get('rating', 'N/A')
            address = place['formatted_address']
            location = place['geometry']['location']
            data.append({'Name': name, 'Rating': rating, 'Address': address, 'Location': location})
        time.sleep(2)

    # Create a DataFrame from the collected data and concat to df
    new_df = pd.DataFrame(data)
    df = pd.concat([new_df,df.loc[:]]).reset_index(drop=True)
    
    return df
```

```python
# Jakarta
for city in cities_jakarta:
    df = scrape_hotels_in_city(df, gmaps, city)

df
```
